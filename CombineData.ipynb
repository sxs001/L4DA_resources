1. Set up

```python
# Importing library
import pandas as pd
# Set numbers for how many rows and columns will be displayed
pd.set_option('display.min_rows', 10) #default will be 10
pd.set_option('display.max_columns', 20)
```

2. Loading Excel
 Formats Into a Pandas Data Frame

 The first Excel sheet contains the details about the items in each order
 The second Excel Sheet contains the details about the reps.


```python
url='https://github.com/tdmhub/L4DA_resources/raw/main/OrderItemV1.xlsx'
df_Item = pd.read_excel(url, sheet_name='Sheet1')
df_Item

```

```python
url='https://github.com/tdmhub/L4DA_resources/raw/main/OrderRepV1A.xlsx'
df_Rep = pd.read_excel(url, sheet_name='Sheet1')
df_Rep
```

Combine the Data Frames  join='outer'
The common column is Order number, so the join is made on this column.



```python
df_ItemRep= pd.merge(df_Item,df_Rep, on='Order No')
df_ItemRep
```

Drop rows with NAN

```python
df_ItemRep2=df_ItemRep.dropna()
df_ItemRep2
```

Create a data frame holding the data for all the reps in the East region and clean and process this data to find the total sales for each rep.


```python
df_ItemRepEast=df_ItemRep2.where(df_ItemRep2.Region=='East')
df_ItemRepEast2=df_ItemRepEast.dropna()
df_ItemRepEast2
```

```python
df_ItemRepEast3=df_ItemRepEast2.pivot(index='Order No',columns='Rep',values='Total')
df_ItemRepEast3
```

```python
df_ItemRepEast4=df_ItemRepEast3.fillna(0)
df_ItemRepEast4
```

```python
df_Eastsum=df_ItemRepEast4.sum()
df_Eastsum
```

Repeat the processing above for the West and Central regions to find the total sales for each rep in these regions.

```python
df_ItemRepWest=df_ItemRep2.where(df_ItemRep2.Region=='West')
df_ItemRepWest2=df_ItemRepWest.dropna()
df_ItemRepWest2
```

```python
df_ItemRepCentral=df_ItemRep2.where(df_ItemRep2.Region=='Central')
df_ItemRepCentral2=df_ItemRepCentral.dropna()
df_ItemRepCentral2
```

You are now to append the three saparate tables for East, West and Central together.

```python
df_Original=pd.concat([df_ItemRepCentral2,df_ItemRepEast2,df_ItemRepWest2])
df_OriginalSorted=df_Original.sort_values('Order No')
df_OriginalSorted
```
