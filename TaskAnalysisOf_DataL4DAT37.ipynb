```python
# Importing library
import pandas as pd
# Set numbers for how many rows and columns will be displayed
pd.set_option('display.min_rows', 10) #default will be 10
pd.set_option('display.max_columns', 20)
url='https://github.com/tdmhub/L4DA_resources/raw/main/online_store_customer_data.csv'
df_Store = pd.read_csv(url)
df_Store
```

```python
df_Store.isna().sum().sort_values(ascending=False)
```

```python
df_Store_NA=df_Store.dropna(subset=["Amount_spent"],inplace=True)
df_Store
```

```python
df_Store.isna().sum().sort_values(ascending=False)
```

```python
df_Store_NA_Gender=df_Store.copy()
df_Store_NA_Gender.dropna(subset=["Gender"],inplace=True)
```

```python
df_Store_NA_Gender.isna().sum().sort_values(ascending=False)
```

```python
df_Store_NA_Employees_status=df_Store.copy()
df_Store_NA_Employees_status.dropna(subset=["Employees_status"],inplace=True)
df_Store_NA_Employees_status.isna().sum().sort_values(ascending=False)
```
